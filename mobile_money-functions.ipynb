A function to determine the number of options that a column has if they valid or not.

```python
# Function to determine number of options
def isvalid(num, start, end):
    '''validate option chosen if within given range: num = number of option, start = first option number, end = last option number'''
    if start == -1:
        if num >= 1 and num <= end or num == start:
            return True
        else:
            return False
    else:
        if num >= start and num <= end:
            return True
        else:
            return False
```

### 3. Create appropriate graphs to visually represent the relationship between financial services accessed (non, momible, both) and age, gender, marital status, land ownership and type of income.

Filtering the DataFrame into those whose use the mobile money service only

```python
momible = df[(df.mobile_money_classification == 2) | (df.mobile_money_classification == 3)]
```

Importing the seaborn module for visualization

```python
import seaborn as sns
```

Classifying the rows in the DataFrame for non mobile money users (non) and mobile money users (momible) in a new column

```python
classification = []
for i in df['mobile_money_classification']:
    if i == 0 or i == 1:
        classification.append('non')
    else:
        classification.append('momible')
df['classification'] = classification
```

##### Visualizing non, momible and both (non + momible) and age

```python
# Age
sns.catplot(y='Q1',kind='count', col='classification', height=14, aspect=.6, data=df).set(xlabel='Age')
sns.catplot(y='Q1',kind='count', height=14,aspect=.6, data=df).set(title='Both', xlabel='Age')
```

There are more users between the age of 20 and 30 who use mobile money services compared to those who do not. With their small numbers there are more elderly users who do not use mobile money services from age 60 and older.

##### Visualizing non, momible and both (non + momible) and gender

```python
# Gender
sns.catplot(x='Q2',kind='count', col='classification', aspect=.6, data=df).set(xlabel='Gender')
sns.catplot(x='Q2',kind='count',aspect=.6, data=df).set(title='Both', xlabel='Gender')
```

1 = Male      
2 = Female

For non mobile money users there more females that do use the service compared to men and for those users who do use the service the number of users that are male and female are the equal. There are more females who do not use the service compared to females that do and less males that do not use the service than those who do.

##### Visualizing the relationship between non, momible and both (non + momible) and type of income
